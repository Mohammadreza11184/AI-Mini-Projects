```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cancer.csv')
```

```python
data = data.drop(['id'],axis=1)
data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
```

```python
data.diagnosis
```

```python
#or
data = data.drop(['id'],axis=1)
c = 0
for k in data.diagnosis:
    if k == 'M':    
        data.diagnosis[c] = 1
    else:
        data.diagnosis[c] = 0    
    c = c + 1     
```

```python
y = data.diagnosis
x0 = data.drop(['diagnosis'],axis=1) 
```

```python
#d = np.array([10,20,30,200,100,50])
```

```python
x = (x0 - np.min(x0)) / (np.max(x0) - np.min(x0))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=42)
```

```python
from sklearn import linear_model 
robot = linear_model.LogisticRegression()

robot.max_iter = 100000
robot.fit(X_train,y_train)
```

```python
t = robot.predict(X_test)

v = y_test == t
v = v.reset_index()
v = v.diagnosis
n = 0
for i in v:
    if i == False:
        n = n + 1
dr = len(v) - n        
darsad = (dr * 100) / len(v)
print(n,'/',len(v)) 
print(darsad)       
```
