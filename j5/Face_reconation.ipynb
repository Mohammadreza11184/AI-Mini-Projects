```python
import face_recognition  as fr
import numpy as np
import cv2
```

```python
img_t = fr.load_image_file('Mohammadreza.jpg')
img_t2 = fr.load_image_file('Alireza.jpg')
img_t3 = fr.load_image_file('Amir Ali.jpg')
img_t4 = fr.load_image_file('Beniamin.jpg')
img_t5 = fr.load_image_file('Sobhan.jpg')
sabt_soorat = fr.face_encodings(img_t)[0]
sabt_soorat2 = fr.face_encodings(img_t2)[0]
sabt_soorat3 = fr.face_encodings(img_t3)[0]
sabt_soorat4 = fr.face_encodings(img_t4)[0]
sabt_soorat5 = fr.face_encodings(img_t5)[0]
```

```python
cam = cv2.VideoCapture(0)
while True:
    _,img = cam.read()
    img = cv2.flip(img,1)
    img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    locate = fr.face_locations(img2)
    if len(locate) > 0 :      
        img = cv2.rectangle(img,(locate[0][3],locate[0][0]),(locate[0][1],locate[0][2]),(0,255,0),3)

        encod = fr.face_encodings(img2)

        result = fr.compare_faces([sabt_soorat,sabt_soorat2,sabt_soorat3,sabt_soorat4,sabt_soorat5],encod[0]) 
        if result[0] == True:
            cv2.putText(img,"Mohammadreza",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
        elif result[1] == True:
            cv2.putText(img,"Alireza",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
        elif result[2] == True:
            cv2.putText(img,"Amir Ali",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)    
        elif result[3] == True:
            cv2.putText(img,"Beniamin",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
        elif result[4] == True:
            cv2.putText(img,"Sobhan",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)        
        else:
             cv2.putText(img,"*Unknown*",(locate[0][3] + 6,locate[0][0] - 6) , cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)        
    cv2.imshow('img',img)
    if cv2.waitKey(1) & 0xFF == ord('p'):
        break
    
cam.release()
cv2.destroyAllWindows()   
```
